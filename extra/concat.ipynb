```python
import pandas as pd
from astropy.table import Table, vstack
import glob
import numpy as np
```

```python
%%bash
# source /arc/home/aydanmckay/ml/xmatch-aydan/proper-xmatching/spectrolabels-cdsmatch.bash
# source /arc/home/aydanmckay/ml/xmatch-aydan/proper-xmatching/outerjoin_spec_x_xpsummary.bash
```

```python
apoeso = Table.read('/arc/home/aydanmckay/intermediarytables/union-apoeso-xpsummary-table.fits').to_pandas()
galleg = Table.read('/arc/home/aydanmckay/intermediarytables/union-galleg-xpsummary-table.fits').to_pandas()
ravseg = Table.read('/arc/home/aydanmckay/intermediarytables/union-ravseg-xpsummary-table.fits').to_pandas()
```

```python
source_ids = []
for i in range(len(apoeso)):
    if pd.isna(apoeso['Source_apo'][i]) == False:
        source_ids.append(apoeso['Source_apo'][i])
    else:
        source_ids.append(apoeso['Source_eso'][i])
apoeso['Source'] = source_ids
source_ids = []
for i in range(len(galleg)):
    if pd.isna(galleg['Source_gal'][i]) == False:
        source_ids.append(galleg['Source_gal'][i])
    else:
        source_ids.append(galleg['Source_leg'][i])
galleg['Source'] = source_ids
source_ids = []
for i in range(len(ravseg)):
    if pd.isna(ravseg['Source_rav'][i]) == False:
        source_ids.append(ravseg['Source_rav'][i])
    else:
        source_ids.append(ravseg['Source_seg'][i])
ravseg['Source'] = source_ids
```

```python
apoeso2 = Table.from_pandas(apoeso)
galleg2 = Table.from_pandas(galleg)
ravseg2 = Table.from_pandas(ravseg)
```

```python
apoeso2.write('/arc/home/aydanmckay/intermediarytables/apoeso-xpsummary-table2.fits', overwrite=True)
galleg2.write('/arc/home/aydanmckay/intermediarytables/galleg-xpsummary-table2.fits', overwrite=True)
ravseg2.write('/arc/home/aydanmckay/intermediarytables/ravseg-xpsummary-table2.fits', overwrite=True)
```
