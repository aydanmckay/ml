## Performing all the xmatches to obtain the huge catalogue of spectroscopic labels.

```python
import pandas as pd
from astropy.table import Table, vstack
import glob
import numpy as np
```

For every bash shell, the bash scripts performed are xmatching through stilts. See the files themselves in their respective folders for the matches that they are performing

```python
%%bash
# source /arc/home/aydanmckay/ml/xmatch-aydan/proper-xmatching/spectrolabels-cdsmatch.bash
# source /arc/home/aydanmckay/ml/xmatch-aydan/proper-xmatching/outerjoin_spec_x_paramp.bash
```

```python
apoeso = Table.read('/arc/home/aydanmckay/intermediarytables/union-apoeso-paramp-table.fits').to_pandas()
galleg = Table.read('/arc/home/aydanmckay/intermediarytables/union-galleg-paramp-table.fits').to_pandas()
ravseg = Table.read('/arc/home/aydanmckay/intermediarytables/union-ravseg-paramp-table.fits').to_pandas()
```

Merging all the source_ids into the same column for the next round of matching.

```python
source_ids = []
for i in range(len(apoeso)):
    if pd.isna(apoeso['Source_apo'][i]) == False:
        source_ids.append(apoeso['Source_apo'][i])
    else:
        source_ids.append(apoeso['Source_eso'][i])
apoeso['Source'] = source_ids
source_ids = []
for i in range(len(galleg)):
    if pd.isna(galleg['Source_gal'][i]) == False:
        source_ids.append(galleg['Source_gal'][i])
    else:
        source_ids.append(galleg['Source_leg'][i])
galleg['Source'] = source_ids
source_ids = []
for i in range(len(ravseg)):
    if pd.isna(ravseg['Source_rav'][i]) == False:
        source_ids.append(ravseg['Source_rav'][i])
    else:
        source_ids.append(ravseg['Source_seg'][i])
ravseg['Source'] = source_ids
```

```python
apoeso2 = Table.from_pandas(apoeso)
galleg2 = Table.from_pandas(galleg)
ravseg2 = Table.from_pandas(ravseg)
```

```python
apoeso2.write('/arc/home/aydanmckay/intermediarytables/apoeso-paramp-table2.fits', overwrite=True)
galleg2.write('/arc/home/aydanmckay/intermediarytables/galleg-paramp-table2.fits', overwrite=True)
ravseg2.write('/arc/home/aydanmckay/intermediarytables/ravseg-paramp-table2.fits', overwrite=True)
```

```python
%%bash
# source /arc/home/aydanmckay/ml/xmatch-aydan/proper-xmatching/dr3xspec.bash
```

Now combining all the individually matched tables through pandas vstack

```python
dr3tabs = glob.glob('/arc/home/aydanmckay/dr3tables/*')
```

```python
tables = []
for it,tab in enumerate(dr3tabs):
    data = Table.read(tab)
    if len(data) == 0:
        print(it,'zero row table')
    else:
        tables.append(data)
```

```python
combined = vstack(tables)
```

```python
combined.write('/arc/home/aydanmckay/input_catalogue_fullset_paramp.fits', overwrite=True)
```
